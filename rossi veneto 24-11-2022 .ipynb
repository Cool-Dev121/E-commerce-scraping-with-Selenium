```python
import selenium
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
#create master dataframe

df = pd.DataFrame(columns=['Wine_name','Producer','Price']) 
```

```python
#Opening Chrome browser
wbD=wb.Chrome('chromedriver.exe')

#Opening webpage
wbD.get('https://www.tannico.it/i-rossi/regioni/veneto.html')


time.sleep(2)  # Allow 2 seconds for the web page to open

scroll_pause_time = 1 #set time
screen_height = wbD.execute_script("return window.screen.height;")   # get the screen height of the web
i = 1

# setting the while loop for the scraping and labeling of extracted data

while True:
    # scroll one screen height each time
    wbD.execute_script("window.scrollTo(0, {screen_height}*{i});".format(screen_height=screen_height, i=i))  
    i += 1
    time.sleep(scroll_pause_time)
    # update scroll height each time after scrolled, as the scroll height can change after we scrolled the page
    scroll_height = wbD.execute_script("return document.body.scrollHeight;")  
    # Break the loop when the height we need to scroll to is larger than the total scroll height
    if (screen_height) * i > scroll_height:
        break

# scraping brands

multiple_brands = wbD.find_elements(By.CLASS_NAME, "productItem__brand")

brand = []

for brands in multiple_brands:
    brand.append((brands).text)

# scraping Wine name

multiple_titles = wbD.find_elements(By.CLASS_NAME, "productItem__title")

title = []

for titles in multiple_titles:
   # list(filter(None, title))
    title.append((titles).text)

# scraping price

multiple_prices = wbD.find_elements(By.XPATH, "//span[@class='new-price']")


price = []

for prices in multiple_prices:
    price.append((prices).text)

#some data modeling is required. The first element extracted in title is empty "" but no data is missing. I am just filtering none value, the df checked afterwards has no problems

title = list(filter(None, title))

num_page_items = len(title)
for i in range(num_page_items):
    try:
        df = df.append({'Wine_name': title[i], 'Producer': brand[i], 'Price': price[i]}, ignore_index=True)
    except IndexError:
        df.append(None)

```
