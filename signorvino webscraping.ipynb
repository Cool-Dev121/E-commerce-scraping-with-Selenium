```python
import selenium
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, WebDriverException
import pandas as pd
import time
import requests
from datetime import datetime as dt

```

## ROSSO

```python
#Opening Chrome browser
wbD=wb.Chrome('chromedriver.exe')


url = 'https://www.signorvino.com/it/vini/rossi/?start=0&sz=5000'
    
#Opening webpage
wbD.get(url)

time.sleep(5)  # Allow the web page to open

# close the cookie window in order to navigate the website

wbD.find_element(By.XPATH, "//*[@id='onetrust-close-btn-container']/button").click()

name = []
price = []
brand = []
variant = []
region = []
year = []
abbinamento = []
bottle = []


elements = wbD.find_elements(By.CLASS_NAME, 'analytics.product.hidden')

print("articles in the page:", len(elements))
for element in elements:
    name.append(element.get_attribute("data-name"))
    price.append(element.get_attribute("data-price"))
    brand.append(element.get_attribute("data-brand"))
    variant.append(element.get_attribute("data-variant"))
    region.append(element.get_attribute("data-dimension3"))
    year.append(element.get_attribute("data-dimension4"))
    abbinamento.append(element.get_attribute("data-dimension6"))
    bottle.append(element.get_attribute("data-dimension42"))

# check if the extracted data have different shape

print("wines in the page:", len(name))
print("prices in the page:", len(price))
print("brand in the page:", len(brand))
print("variants in the page:", len(variant))
print("regions in the page:", len(region))
print("years in the page:", len(year))
print("abbinamenti in the page:", len(abbinamento))
print("bottles in the page:", len(bottle))

#create dataframe


red = pd.DataFrame(name ,columns=['Wine_name'])
red["Price"] = price
red["Producer"] = brand
red["Wine_Type"] = variant
red["Region"] = region
red["Year"] = year
red["Food_Match"] = abbinamento
red["Bootle_Type"] = bottle
red["Category"] = "Rosso"


mask = '%d%m%Y'
dte = dt.now().strftime(mask)
fname = "Catalogo_signorvino_rosso{}.xlsx".format(dte)
red.to_excel(fname)
```
